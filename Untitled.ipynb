```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ssq20058.txt",header=None) #,names=['period','red_1','red_2','red_3','red_4','red_5','red_6','blue_1']
df = df[0].str.split('，', expand=True)
df.columns = ['period','red_1','red_2','red_3','red_4','red_5','red_6','blue_1']
df.iloc[:,1:] = df.iloc[:,1:].astype('int')

data = []
ind = []
for i in range(1,34):
    ind.append(i)
    data.append(len(df.iloc[:,1:-1][df.iloc[:,1:-1]==i].dropna(how='all')))
sr = pd.Series(data,index=ind).sort_values(ascending=False)
sr.plot(kind='bar', figsize=(10,5), title='Red Ball, 2003-2020', grid=True, ylim=(400,520)) # , yticks=np.arange(160,250,10)
```

```python
data = []
ind = []
for i in range(1,17):
    ind.append(i)
    data.append(len(df.iloc[:,-1:][df.iloc[:,-1:]==i].dropna(how='all')))
sr = pd.Series(data,index=ind).sort_values(ascending=False)
sr.plot(kind='bar', figsize=(10,5), title='Blue Ball, 2003-2020', grid=True, ylim=(130,180)) #yticks=np.arange(160,250,10),
```
