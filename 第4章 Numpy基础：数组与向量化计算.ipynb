上面创建一个正态分布随机数组（均值为0，方差为1）。下面对其进行了验算。

```python
mean = np.mean(data)
variance = np.std(data) ** 2
print(mean, variance)
```

如何生成指定均值和方差的随机数组呢？下面的方差市sigma的平方。

sigma * np.matlib.randn(....) + mu

```python
import numpy.matlib

random3 = 2.5 * np.matlib.randn((1000,1000)) + 3
mean3 = np.mean(random3)
variance3 = np.std(random3) ** 2
print(mean3, variance3)
```
