```python
import tushare as ts
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

使用tushare包获取某股票的历史数据

```python
df = ts.get_k_data("600519",start='1990-01-01')
df.to_csv("600519.csv")
```

```python
df = pd.read_csv("600519.csv", index_col='date', parse_dates=['date'])[['open','close','high','low']]
df
```

输出该股票所有收盘价比开盘价上涨3%的日期

```python
df[(df['close']-df['open'])/df['open']>=0.03].index
```

输出所有开盘比前日收盘跌幅超过2%的日期     shift()函数

```python
df[(df['open'] - df['close'].shift(1))/df['close'].shift(1)<=-0.02].index
```

假如我从2010年1月1日开始，，每月第一个交易日买入一首股票，每年最后一个交易日卖出所有股票，到今天为止，我的收益如何？

```python
df
```

```python
price_last = df['open'][-1]
df = df['2001-09':'2020-06'] #剔除收尾无用的数据
df
```

```python
df_monthly = df.resample("M").first() #每月第一天的数据表
df_yearly = df.resample("A").last()[:-1] #最后一年年底还没到，所以去掉
```

```python
cost_money = 0
hold = 0
for year in range(2001, 2021):
    cost_money += df_monthly[str(year)]['open'].sum() * 100
    hold += len(df_monthly[str(year)]['open']) * 100
    if year != 2020:
        cost_money -= df_yearly[str(year)]['open'][0] * hold
        hold = 0
    print("cost_money in " + str(year) + ":" , cost_money)    
    
cost_money -= hold * price_last

print(-cost_money)
```
