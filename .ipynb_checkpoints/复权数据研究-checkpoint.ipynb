```python
df_bfq = pd.read_csv("601888_None.csv", index_col='date', parse_dates=['date']) # 不复权数据
df_qfq = pd.read_csv("601888_qfq.csv", index_col='date', parse_dates=['date']) # 前复权数据
df_hfq = pd.read_csv("601888_hfq.csv", index_col='date', parse_dates=['date']) # 后复权数据
```

```python
df_bfq['qfq'] = df_qfq['close']
df_bfq['hfq'] = df_hfq['close']
df_bfq.rename(columns = {"close": "bfq"}, inplace=True)
df_bfq['qfq-bfq'] = df_bfq['qfq'] - df_bfq['bfq']
df_bfq['delta'] = 0.52 * df_bfq['bfq']/df_bfq['bfq']['2018-06-21']  # 2018-06-22:10派5.20元
```
