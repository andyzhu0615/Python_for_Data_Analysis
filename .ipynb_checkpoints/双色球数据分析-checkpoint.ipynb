```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ssq.csv", index_col='开奖日期', parse_dates=['开奖日期'])
# df['中奖号码'].str.split('  ', expand=True).columns(['A','B','C','D','E','F','G'])
df1 = df['中奖号码'].str.split('  ', expand=True)

df1.columns = ['red1','red2','red3','red4','red5','red6','blue']

df1.index = df.index

# df1[['red1','red2','red3','red4','red5','red6']] = df1[['red1','red2','red3','red4','red5','red6']].astype('int')
df1 = df1.astype('int')

data = []
ind = []
for i in range(1,34):
    ind.append(i)
    data.append(len(df1.iloc[:,:6][df1==i].dropna(how='all')))
sr = pd.Series(data,index=ind)
sr.plot(kind='bar', figsize=(10,5), title='Red Ball', yticks=np.arange(160,250,10), grid=True, ylim=(150,250)) #
```

```python
data = []
ind = []
for i in range(1,17):
    ind.append(i)
    data.append(len(df1.iloc[:,-1:][df1==i].dropna(how='all')))
sr = pd.Series(data,index=ind)
sr.plot(kind='bar', figsize=(10,5), title='Red Ball',  grid=True, ylim=(0,100)) #yticks=np.arange(160,250,10),
```
