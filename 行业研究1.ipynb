```python
import tushare as ts
import matplotlib.pyplot as plt

df=ts.get_hist_data('601857',start='2016-06-15',end='2019-11-06')

#复权类型，qfq-前复权 hfq-后复权 None-不复权，默认为qfq
df1=ts.get_k_data('601857',start='2015-06-15',end='2019-11-06')#前复权
df2=ts.get_k_data('601857',start='2015-06-15',end='2019-11-06',autype='hfq')#后复权
df3=ts.get_k_data('601857',start='2015-06-15',end='2019-11-06',autype='None')#不复权
```
